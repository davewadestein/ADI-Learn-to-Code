# `while` loop example

````python
import random # "batteries included"
# what do you think the line below does?
my_number = random.randint(1, 100)
guess = 0 
# loop until...?

while guess != my_number:
    guess = int(input('Enter your guess (0 to give up): '))
    if guess == 0:
        print("Sorry that you're giving up!")
        break # abnormal termination
    elif guess > my_number:
        print("Guess was too high")
    elif guess < my_number:
        print("Guess was too low")        
else:
    print("Congratulations. You guessed it!")
# break would put us here
````

* Write a Python program to read in a list of items possibly containing duplicates, and then constructs a new list which contains the elements from the original list, with the order preserved, but the duplicates removed


````python
# 1a. read a line of input
# 1b. turn it into a list (.split()) ... split a string into a list
# 2. create a new (empty) list to hold the non-dupes
# 3. for each word in the original list:
#    4. if word is not in the new list:
#         5. add it to the new list
words = input('Enter some words and I will abracadabra: ').split() # 1a and 1b
````

````python
words # check our work
````

````python
non_dupe_words = [] # step 2
for word in words: # step 3
    if word not in non_dupe_words: # step 4
        non_dupe_words.append(word) # step 5

non_dupe_words
````

````python
# 0. create an empty list
# 1. read words from the keyboard until 'quit'
# 2. if word begins with '-', remove it from the list
#    2a. (extra) split up the words on the line and remove each individually
# 3. if the word is '-', reverse the list
# 4. otherwise, add the word to the list
#.   4a. (extra) split the line into words and add each word to the list
# 5. print the list
````

````python
# first cut, we can do better...
# 1. no error checking
# 2. what if user just hits return?
# 3. add additional extras from above
list_of_words = []

while True: # step 1...
    word = input('Enter a word: ')
    if word == 'quit':
        break # also step 1
    # step 2...
    if word.startswith('-'): # - is the first letter
        if word == '-': # ...and in fact the only letter
            # step 3
            list_of_words = list_of_words[::-1] # .reverse() is perhaps better
        else: # begins with -, but not just a -
            # also step 2
            list_of_words.remove(word[1:]) # ignore '-', as it's not part of the word
    else: # does not start with -, so step 4
        list_of_words.append(word)
    print(list_of_words)
````

````python
# second cut

list_of_words = []

while True: # step 1...
    response = input('Enter a word: ')
    if not response: # response == ''
        continue
    if response == 'quit':
        break # also step 1
    # step 2...
    if response.startswith('-'): # - is the first letter
        if response == '-': # ...and in fact the only letter
            # step 3
            if len(list_of_words) < 2:
                print('Hey, trying to reverse fewer than 2 items is moon man talk!')
            else:
                list_of_words = list_of_words[::-1] # .reverse() is perhaps better
        else: # begins with -, but not just a -
            # also step 2
            for word in response[1:].split(): # split everything AFTER the hyphen/dash/minus
                if word in list_of_words:
                    list_of_words.remove(word) # ignore '-', as it's not part of the word
                else:
                    print(f'Pay attention! "{word}" is not in the list!')          
    else: # does not start with -, so step 4
        words = response.split() # list of 1+ or more words
        list_of_words.extend(words)
    print(list_of_words)
````
